```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Analyse des données sur les gravités des accidents depuis 2012.
#### Nous sommes en mesure de voir que malgré un nombre très élevés d'accident les accidents graves sont minime par rapport
#### au autres accidents.

```python
def obtenir_gravite_accident(donnees_accident):
    
    #Obtenir le nombre d'accident par gravité des données extraite du csv
    data_gravite_accident = donnees_accident["GRAVITE"].value_counts()
    #création du diagrame a bar
    ax = data_gravite_accident.plot(kind="bar", figsize=(20,10), color="#1565c0", fontsize=14)
    #Modification du titre, le nom des axes et la légendes
    ax.set_title("Le nombre d'accidents en fonction des gravités de ceux-ci depuis 2012", fontsize=18)
    plt.xlabel('Gravité de l\'accident', fontsize=18)
    plt.ylabel('Nombres d\'accidents', fontsize=18)
    plt.legend(['Nombres d\'accients'])

    #Ajout du nombre d'accident en haut de chaque gravité pour bien voir le nombre de ceux-ci
    for index,data in enumerate(data_gravite_accident):
        plt.text(x=index-.15, y=data+4 , s=f"{data}" , fontdict=dict(fontsize=20))

```

```python
#Lecture du fichier CSV
donnees_accident = pd.read_csv("accidents_2012_2018.csv", delimiter=",", parse_dates=["DT_ACCDN"])
numbre_accident = donnees_accident['AN'].value_counts().sum()
print(f'Il y a eu {numbre_accident} d\'accidents depuis 2012 à Montréal')
#Analyse de la première donnée - Gravité des accidents depuis 2012.
obtenir_gravite_accident(donnees_accident)
```
