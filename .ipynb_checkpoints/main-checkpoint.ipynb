```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Lecture du fichier CSV et calcul du nombre d'accident
donnees_accident = pd.read_csv("accidents_2012_2018.csv", delimiter=",", parse_dates=["DT_ACCDN"])
numbre_accident = donnees_accident['AN'].value_counts().sum()
print(f'Il y a eu {numbre_accident} d\'accidents depuis 2012 à Montréal')
```

# Analyse des données sur le genre d'accident depuis 2012.

```python
def obtenir_gravite_accident(donnees_accident):    
    #Obtenir le nombre d'accident par gravité des données extraite du csv
    data_gravite_accident = donnees_accident["GRAVITE"].value_counts()
    #création du diagrame a bar
    ax = data_gravite_accident.plot(kind="bar", figsize=(20,15), color="#1565c0", fontsize=14)
    #Modification du titre, le nom des axes et la légendes
    ax.set_title("Le nombre d'accidents en fonction des gravités de ceux-ci depuis 2012", fontsize=18)
    plt.xlabel('Gravité de l\'accident', fontsize=18)
    plt.ylabel('Nombres d\'accidents', fontsize=18)
    plt.legend(['Nombres d\'accients'])

    #Ajout du nombre d'accident en haut de chaque gravité pour bien voir le nombre de ceux-ci
    for index,data in enumerate(data_gravite_accident):
        plt.text(x=index-.15, y=data+4 , s=f"{data}" , fontdict=dict(fontsize=20))
```

```python
# Permet d'obtenir le titre de chaque titre du fichier csv créé manuellement
def obtenir_titre_genre():
    #dictionnaire genre avec clé code : valeur= genre
    dictionnaire_genre = {}  
    donnees_genre = pd.read_csv("genre.csv", delimiter=",")
    # Création du dictionnaire
    for index, row in donnees_genre.iterrows():
        if row['genre'] in dictionnaire_genre.keys():
            pass
        else:
            dictionnaire_genre[row['genre']] = row["titre"]
    return dictionnaire_genre 
       
# Fonction pour obtenir et afficher le tableau des genre d'accident
def obtenir_genre_accident(donnees_accident):
    dictionnaire_genre = obtenir_titre_genre();
    #Obtenir le nombre d'accident par genre des données extraite du csv
    data_genre_accident = donnees_accident["CD_GENRE_ACCDN"].value_counts()
    
    #Pour chaque données du fichier excel avec un genre Int, on va chercher son titre avec le dictionnaire de genre
    titre = []
    total_accident = []
    for genre, total in data_genre_accident.iteritems():
        if genre in dictionnaire_genre.keys():
            titre.append(dictionnaire_genre[genre])
            total_accident.append(total)
    
    # Création d'un dataframe pour faire l'affichage du tableau 
    pd_genre_accident = pd.DataFrame({ 'Genre':titre, 'Total': total_accident})
    
    #création du diagrame a bar
    ax = pd_genre_accident.plot(kind="barh", figsize=(20,15), color="#ff5722", fontsize=14)
    
    #Modification du titre, le nom des axes et la légendes
    ax.set_title("Le nombre d'accidents en fonction du genre de ceux-ci depuis 2012", fontsize=18)
    plt.xlabel('Nombres d\'accidents', fontsize=18)
    plt.ylabel('Genre d\'accidents', fontsize=18)
    plt.legend(['Genre d\'accidents'])
    
    # Permet de mettre les genre en string pour remplacé les type du fichier excel.
    ax.set_yticklabels(pd_genre_accident["Genre"],fontsize=14)

    #Ajout du nombre d'accident en haut de chaque genre pour bien voir le nombre de ceux-ci
    for index,data in enumerate(data_genre_accident):
        plt.text(x=data+1, y=index-0.2 , s=f"{data}" , fontdict=dict(fontsize=14))

```

```python
#Analyse de la deuxième donnée - Genre des accidents depuis 2012.
obtenir_genre_accident(donnees_accident)
    
```

# Analyse des données sur les gravités des accidents depuis 2012.

```python
#Analyse de la première donnée - Gravité des accidents depuis 2012.
obtenir_gravite_accident(donnees_accident)
```

# Extraction du pourcentage d'accident par années

```python
def afficher_repartition_accident_selon_annee(donnees_accident):
    annee = [2012,2013,2014,2015,2016,2017,2018]
    nombre_accident_par_annee = [
        donnees_accident[donnees_accident['AN'] == 2012].count()['AN'],
        donnees_accident[donnees_accident['AN'] == 2013].count()['AN'],
        donnees_accident[donnees_accident['AN'] == 2014].count()['AN'],
        donnees_accident[donnees_accident['AN'] == 2015].count()['AN'],
        donnees_accident[donnees_accident['AN'] == 2016].count()['AN'],
        donnees_accident[donnees_accident['AN'] == 2017].count()['AN'],
        donnees_accident[donnees_accident['AN'] == 2018].count()['AN'],        
    ]

    # Ex pris de https://medium.com/@kvnamipara/a-better-visualisation-of-pie-charts-by-matplotlib-935b7667d77f
    # Création du Pie chart qui représente les répartition des accidents selon les années
    fig, ax = plt.subplots()
    ax.pie(nombre_accident_par_annee, labels=annee, autopct='%1.1f%%', shadow=False, startangle=95)
    # Equal aspect ratio ensures that pie is drawn as a circle
    ax.axis('equal')
    ax.set_title("Pourcentage du nombre d\'accident par années'")
    plt.tight_layout()

afficher_repartition_accident_selon_annee(donnees_accident)
    
```
